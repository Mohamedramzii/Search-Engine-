## Preprocessing

```python
file = open("stopwords/stopwords.txt","r")
stop_words = []
for line in file:
    line_strip = line.strip()
    line_split= line.split()
    stop_words = stop_words +line_split
file.close()
#print (stop_words)
```

```python
Stop_word_i=[]
def preprocess(text):
    processed_text=np.char.lower(text)
    symbols = "!\"#$%&()*+-./:;<=>?@[\]^_`{|}~\n"
    for i in range(len(symbols)):
        processed_text = np.char.replace(processed_text, symbols[i], ' ')
        processed_text = np.char.replace(processed_text, "  ", " ")
    processed_text = np.char.replace(processed_text, ',', '')
    processed_text=np.char.replace(processed_text, "'", "")
    
    words = word_tokenize(str(processed_text))
    new_text = ""
    
    for w in words:
        if w not in stop_words and len(w) > 1:
            new_text = new_text + " " + w
        else:  Stop_word_i.append(w) 
    return new_text
```

```python
# Calculate Document Frequency df : occurance of term in N of documents
DF = {}
for i in range(N):
    tokens = processed_text[i]
    for w in tokens:
        try:
            DF[w].add(i)
        except:
            DF[w] = {i}
for i in DF:
    DF[i] = len(DF[i])
```

### Calculate  Term Frequency and IDF 

```python
import math
tf_idf = {}
term_freq={}
w_term_freq={}
IDF={}

for i in range(N):
    tokens = processed_text[i]    
    freq = Counter(tokens)
    
    for token in np.unique(tokens):
        tf=1+math.log(freq[token])
        term_freq[i, token]=freq[token]
        w_term_freq[i,token]=tf
        idf =  math.log(N/DF[token],10)
        IDF[token]=idf
#         print(token+' in document '+str(i)+'\n\tfrequancy count is : '+str(freq[token])+
#               '\n\tTerm frequency is  :  '+str(tf)+'\n\tIDF is : ' +str(idf))
        tf_idf[i, token] = tf*idf
 
```
